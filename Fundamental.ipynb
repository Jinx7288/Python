### 位运算符

+ 两个二进制表示
+ & 按位与，对应位都为1，结果位为1，否则为0
+ | 按位或，对应位都为0，结果位为0，否则为1
+ ^ 按位异或，对应位相同时为0，否则为1

当两数精度不同时，结果为高精度，低精度补0以计算

+ ~ 按位取反，按位非，0变1，1变0
+ << 左移位运算符，二进制操作数向左移动n位数，高位丢弃，低位补0，相当于乘2的n次幂
+ \>\> 右移位运算符，二进制操作数向右移动n位数，低位丢弃，高位补0，相当于除2的n次幂

```python
~2,2<<9,1024>>9
```

##### while-else for else结构

```python
flag = 39
while flag:
    flag -=1
    if flag == 7:
        break   #一旦跳出循环，else不会执行
else:
    print("循环正常结束")
    print(flag)
for i in range(flag):
    if flag == 7:
        break
else:
    print("循环正常结束")
```
