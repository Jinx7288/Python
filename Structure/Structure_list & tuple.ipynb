#####  列表生成式
1. 直接生成
2. 条件筛选
3. 表达式生成

```python
list0 = [i for i in range(10)]
list1 = [i for i in range(10) if i%2==0]
list2 = [i if i%2==0 else None for i in range(10)]
list3 = [str(i)+"th" for i in range(10)]
list0,list1,list2,list3
```

## tuple

元组的创建，删除，索引，切片，遍历，生成式都同列表，但不能赋值修改，修改必须整个变量全部重新赋值  

但可以通过一些操作实现修改

```python
tuple1=(1,2,3,4,5)
tuple2=(i for index,i in enumerate(tuple1) if index != 0 )
list(tuple2)
```
