编解码

```python
string = "Hello World"
bytes = string.encode(encoding="utf-8",errors="strict")
print(bytes)
```

```python
string = bytes.decode(encoding="utf-8",errors="strict")
print(string)
```

encode编码，decode解码  
encoding：编码格式  
errors  

    1. default，strict，遇到非法字符就抛出异常
    2. ignore，忽略非法字符
    3. replae，用问号代替非法字符
    4. xmlcharrefreplace,使用XML的字符引用

字符串切片

```python
string = "".join([str(i) for i in range(10)])
print(string)
```

string\[start:end:step\],start默认0，end默认len(string)，也就是从start到最后，step默认1

**切片只要记住几个原则**：
1. 先看step正负号，决定遍历方向，同时确定默认为[0::1]或[-1::1]
2. 无论如何，start始终是起点，stop始终是不取的终点
3. 出现逻辑错误，起点在终点后之类，输出为空数组，也即，start-->stop方向和step方向不同时为空
4. 出现范围之外的数字按默认处理
5. 正负索引混用有点坑，判断清楚start-->stop方向

#### 字符串方法

```python
string = "Hello World"
```

```python
string.lower()
```

```python
string.upper()
```

```python
string.capitalize() #将首字母转化为大写，其他小写
```

```python
string.islower()
```

```python
string.isprintable() #判断是否都是可打印的,比如转义字符不能打印
```

```python
string.isnumeric() #判断是否全是数字
```

```python
string.isspace() #判断是否全是空格
```

```python
string.startswith("Hello",0,-1) #以什么开始,可指定范围
```

```python
string.endswith("Hello",0,-1) # 以什么结束，可指定范围
```

```python
string.count("e",0,-1) #计数，可指定范围,不存在返回0,存在返回个数
```

```python
string.find("e",0,-1) #查找，可指定范围，不存在返回-1，存在返回索引，另有 rfind() ，从右开始查找
```

```python
string.index("e",0,-1) #同 find ，但若不存在会抛出异常
```

```python
string.split(" ",maxsplit = -1)  #指定分隔符，默认为空字符，最大分割次数，默认分割全部
```

```python
list(string),[c for c in string] #全部分割，别傻傻地想用split了
```

```python
"".join(list(string)) # 合并字符串,双引号里可以有分隔符，参数为iterable
```

```python
string.replace("e","o",1)  # 替换，前count个
```

```python
string.strip("H \n\t") # 除去字符串两边指定的字符，默认为空字符，可指定多个
```

```python
string.lstrip(),string.rstrip() #仅去除左边/右边，直到第一个没有指定的字符
```

```python
string.center(20," ")  #从中间向两边扩张到指定字符数，默认空格填充，可指定
```

```python
string.zfill(20)  #从左侧补零到指定位数
```

##### 字符串格式化

template.format()

format(\*args)   多个参数，用于渲染进模板
template:     {index : fill align sign # width grouping .precison type}  
index 参数索引，若没有也得写冒号，默认按顺序渲染  
fill 指定填充空的任意字符，默认空格  
align 水平对齐方式，\<左对齐，\>右对齐，\^居中，该属性需和width一起使用  
sign +正数加正号，负数加负号 -正数不变，复数加负号 空格正数加空格，负数加负号  
\# 是否显示进制前缀  
width 宽度  
\.precision 保留的小数位数  
type 类型 还是查吧
