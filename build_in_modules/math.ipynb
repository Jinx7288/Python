### 数论与表示函数

```python
x,y=3,9
```

```python
math.fabs(x)  #绝对值
```

```python
math.fmod(x,y)  #返回x对y的模
```

```python
math.fsum([x,y])  #浮点数精确求和，参数为iterable
```

```python
math.prod([x,y],start=2)  # 对可迭代对象求积，只能是数字，start为自定义因数，默认为 1
```

```python
math.comb(y,x)  # 数学上的组合数，C，无序 y 个里取 x 个
```

```python
math.perm(y,x)  #排列数，A，有序 y 里取 x 个，没有 x 返回 y 的阶乘
```

```python
math.copysign(x,y)  #复制符号，计算|x|*|y|/y, 相当于把 y 的正负号赋给 x
```

```python
math.ceil(x)  #向上取整，返回不小于x的最小整数
```

```python
math.floor(x)  #向下取整，返回不大于x的最大整数
```

```python
math.factorial(x)  #阶乘，x为小数复数抛出ValueError
```

```python
math.gcd(x,y) #返回最大公约数，参数为多个整数
```

```python
math.lcm(x,y)  # 返回最小公倍数 ，参数为多个整数
```

```python
math.frexp(x)  #返回(m,e),满足x == m * 2**e  当x=0，返回(0.0,0)
```

```python
math.ldexp(x,y)  #frexp的逆运算，计算 x * 2 ** y
```

```python
math.modf(3.14)  #返回x的小数和整数部分
```

```python
math.trunc(3.14)  #返回整数部分
```

```python
math.isfinite(x)  #判断是不是无穷和 NaN ,Return True if x is neither an infinity nor a NaN
```

```python
math.isinf(x)  #判断是不是正或负无穷
```

```python
math.isnan(x)  #判断是不是NaN,返回布尔值
```

```python
math.isqrt(x)  #返回 x 的平方根向下取整的结果，使得 a² ≤ x 的最大整数 a。
                #返回使得 n ≤ a² 的最大整数 a，可以用 a = 1 + isqrt(n-1)
```

###  三角函数

```python
math.sin(x)  #返回 x 弧度的正弦值。
```

```python
math.cos(x)  #返回 x 弧度的余弦值。
```

```python
math.tan(x)  #返回 x 弧度的正切值。
```

```python
math.asin(x)  #返回以弧度为单位的 x 的反正弦值。 结果范围在 -pi/2 到 pi/2 之间
```

```python
math.acos(x) #返回以弧度为单位的 x 的反余弦值。 结果范围在 0 到 pi 之间
```

```python
math.atan(x)  #返回以弧度为单位的 x 的反正切值。 结果范围在 -pi/2 到 pi/2 之间
```

```python
math.atan2(y, x) #以弧度为单位返回 atan(y / x) 。结果是在 -pi 和 pi 之间
```

```python
math.dist(p, q)  #返回 p 与 q 两点之间的欧几里得距离，以一个坐标序列（或可迭代对象）的形式给出。 两个点必须具有相同的维度。
#  大致相当于   sqrt(sum((px - qx) ** 2.0 for px, qx in zip(p, q)))
```

```python
math.hypot([x,y])  #返回欧几里得范数，sqrt(sum(x**2 for x in coordinates))。 这是从原点到坐标给定点的向量长度。
```
